$$
\Large \textbf
{
Problema 3
}
$$

$ \textbf{a)} $ Un paracaidista de 70kg salta desde un globo aerostático fijo. El coeficiente de resistencia al aire es de aproximadamente 12.5 kg/seg (considerar la aceleración de la gravedad como 10 $ \frac{m}{s^2} $. Utilizando la expresión exacta evalúe la velocidad del paracaidista cada 2 segundos despues del salto, y calcule su velocidad límite. Grafique el resultado

$ \textbf{b)} $ Repita el gráfico con la solución discreta, tomando intervalos de 2 seg, 1 seg, y 0.5 seg

$ \textbf{c)} $ Haga el mismo cálculo pero suponiendo ahora que la fuerza de rozamiento se debe expresar como $ F=-c'v\ 2 $ (tomar c' = 0.25 kg/m )

$
\text{Ecuación a resolver:}
$
$ $
$ \Large
m\frac{dv}{dt}=mg-\gamma v
$

$
\text{La solución exacta es:}
$
$ $
$ \Large
v=\frac{gm}{\gamma}(1-e^{-(\gamma/m)t})
$

```python
import math
import matplotlib.pyplot as plt
```

```python
def paracaidas(g, m, aire, tiempo, dif_t):
    t = [0]
    v = [0]
    i = dif_t
    while t[-1] <= tiempo:  # tiempo = hasta que tiempo se quiere graficar
        t.append(i)
        v.append(g*m/aire*(1-math.exp(-(aire/m)*t[-1])))
        i += dif_t
    return t, v
```

```python
g = 10  # m/s2
m = 70  # kg
aire = 12.5  # kg/s
```

$ \Large \textbf{a)} $

```python
ta, va = paracaidas(g, m, aire, 60, 2)

fig, ax = plt.subplots()
ax.plot(ta, va, linestyle='dotted')
plt.show()
```

$ \Large \textbf{b)} $

```python
tb0, vb0 = paracaidas(g, m, aire, 60, 2)
tb1, vb1 = paracaidas(g, m, aire, 60, 1)
tb2, vb2 = paracaidas(g, m, aire, 60, 0.5)

fig, ax = plt.subplots(1, 3)
ax[0].plot(tb0, vb0, linestyle='dotted')
ax[0].set_xlabel('t [seg]')
ax[0].set_ylabel('v [m/s]')
ax[1].set_title('saltos de 2seg ')
ax[1].plot(tb1, vb1, linestyle='dotted')
ax[0].set_title('saltos de 1seg ')
ax[2].plot(tb2, vb2, linestyle='dotted')
ax[2].set_title('saltos de 0.5seg ')
plt.show()
```

$ \Large \textbf{c)} $

```python
aire = 2*0.25

tb0, vb0 = paracaidas(g, m, aire, 60, 2)
tb1, vb1 = paracaidas(g, m, aire, 60, 1)
tb2, vb2 = paracaidas(g, m, aire, 60, 0.5)

fig, ax = plt.subplots(1, 3)
ax[0].plot(tb0, vb0, linestyle='dotted')
ax[0].set_xlabel('t [seg]')
ax[0].set_ylabel('v [m/s]')
ax[1].set_title('saltos de 2seg ')
ax[1].plot(tb1, vb1, linestyle='dotted')
ax[0].set_title('saltos de 1seg ')
ax[2].plot(tb2, vb2, linestyle='dotted')
ax[2].set_title('saltos de 0.5seg ')
plt.show()
```
