```python
import numpy as np   # Importo paquete de álgebra lineal
import matplotlib.pyplot as plt   # Importo paquete de gráficos
import mef
plt.rc('figure', figsize=(15,8))   # Para gráficos
plt.rc('font',size=22)             # Para gráficos
from scipy import linalg
```

## Punto 1

Calcule las **frecuencias naturales de oscilación transversal** de una barra de acero (densidad $\rho = 7850\: kg/m^3$ y módulo de elasticidad $E = 210\: GPa$), de $1\: m$ de longitud, $10\: {cm}^2$ de área, momento de inercia de $10\: {cm}^4$. 
- Realice el cálculo para un número variable de elementos y estudie la convergencia de sus resultados.
- Haga el cálculo para matrices de masa concentrada y consistente. 
- Grafique los desplazamientos en función de la posición para los primeros modos normales.

<center><img src="ImgProb01.png"></center>

```python
# Defino los grados de libertad por nodo "glxn".
glxn = 2
```

```python
# Defino el módulo de elasticidad "E" de la viga.
E = 210e9   # Pa

# Defino densidad "rho" del material.
rho = 7850  # kg/m^3

# Defino la longitud total de la viga.
long = 1   # m

# Defino la momento de inercia "I".
I = 1e-7   # m^4

# Defino el área "A" de la barra.
A = 1e-3   # m^2
```

## Cálculo empleando Matriz de masa Consistente

Se creó la función $frec_M_consistente_barra_empotrada$ en el módulo $mef$, ella es quien calcula las frecuencias y desplazamientos en Y asociados a los modos normales transversales.

### Estudio de la CONVERGENCIA, variando el número de elementos $Ne$

Se consideraron números de elementos $Ne = 4$ y $Ne = 100$. Para cada uno, se representaron gráficamente **Dezplazamiento vs. Posición para los modos normales 1, 2, 3 y 4**.

```python
Ne = 8
frec, d = mef.frec_M_consistente_barra_empotrada(E, A, long, I, rho, Ne, glxn)
for i in range(4):   # Grafico los primeros 4 modos normales.
    plt.plot(np.arange(Ne+1)*12.5, d[:, i], label='Modo '+str(i+1)+': f = '+str(round(frec[i], 2))+' Hz')
plt.legend()
plt.grid()
plt.title('Masa Consistente; Ne = 8')
plt.xlabel('Posición en X (cm)')
plt.ylabel('Desplazamiento en Y (cm)')
plt.show()

Ne = 100
frec, d = mef.frec_M_consistente_barra_empotrada(E, A, long, I, rho, Ne, glxn)
for i in range(4):   # Grafico los primeros 4 modos normales.
    plt.plot(np.arange(Ne+1), d[:, i], label='Modo '+str(i+1)+': f = '+str(round(frec[i], 2))+' Hz')
plt.legend()
plt.grid()
plt.title('Masa Consistente; Ne = 100')
plt.xlabel('Posición en X (cm)')
plt.ylabel('Desplazamiento en Y (cm)')
plt.show()
```

**Conclusión**: Al aumentar $Ne$, la solución se suaviza y converge.
